```python
%matplotlib inline

from yolo7 import Yolo7
import cv2
import matplotlib.pyplot as plt


# creando la clase Yolo7

detector_horizontal = Yolo7(weights='/home/pqbas/projects/LABINM_Robotics_Automation/SeedlingsNet/detectors/weights/yolov7-hseed.pt',#'./weights/berries5/weights/best.pt', 
                data='./yolov7/opt.yaml', 
                device='cuda:0')


detector_vertical = Yolo7(weights='/home/pqbas/projects/LABINM_Robotics_Automation/SeedlingsNet/detectors/weights/yolov7-vseed.pt',#'./weights/berries5/weights/best.pt', 
                data='./yolov7/opt.yaml', 
                device='cuda:0')
```

```python
%matplotlib inline

# leyendo la imagen
img = cv2.imread('/home/pqbas/projects/LABINM_Robotics_Automation/SeedlingsNet/classifiers/data/dataset/images/horizontal/rgb/0a3984ef.jpg')
predictions = detector_horizontal.predict(img)

# dibujando el resultado
result = detector_horizontal.plot_prediction(img, predictions)
plt.imshow(result)
plt.show()

# dibujando el resultado de la mascara
mascara = predictions[0].mask
plt.imshow(mascara)
plt.show()
```

```python
%matplotlib inline

# leyendo la imagen
img = cv2.imread('/home/pqbas/projects/LABINM_Robotics_Automation/SeedlingsNet/classifiers/data/dataset/images/vertical/rgb/0bea3c40.jpg')
predictions = detector_vertical.predict(img)

# dibujando el resultado
result = detector_vertical.plot_prediction(img, predictions)
plt.imshow(result)
plt.show()

# dibujando el resultado de la mascara
mascara = predictions[0].mask
plt.imshow(mascara)
plt.show()
```
